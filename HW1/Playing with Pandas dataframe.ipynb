```python
from scipy.io import arff
import pandas as pd
import numpy as np
```

```python
data = arff.loadarff('4year.arff')
df = pd.DataFrame(data[0])
```

### Create a new dataframe with only 4 feataures (and and Bankruptcy). Properly rename the columns to X1, X2, X7, and X10

```python
df1 = df[['bankruptcy', 'Attr1', 'Attr2', 'Attr7', 'Attr10']]
df1.columns = ['Bankruptcy', 'X1', 'X2', 'X7', 'X10']
df1
```

### Fill-in the missing values (nan) with the column means. (Use pd.fillna() or See Ch 4 of PML)

```python
features = ['X1', 'X2', 'X7', 'X10']
df1.fillna({features[i]: df1.mean()[i] for i in range(4)}, inplace = True)
df1
```

### Find the mean and std of the 4 features among all, bankrupt and still-operating companies (3 groups).

```python
mean = df1.groupby(['Bankruptcy']).mean().append(pd.Series(df1.mean()), ignore_index=True).drop(axis = 1, columns = 'Bankruptcy')
mean.index = ['still-operating', 'bankrupt', 'all']
print('The mean of features:')
mean
```

```python
std = df1.groupby(['Bankruptcy']).std().append(pd.Series(df1.mean()), ignore_index=True).drop(axis = 1, columns = 'Bankruptcy')
std.index = ['still-operating', 'bankrupt', 'all']
print('The std of features:')
std
```

### How many companies satisfy the condition, X1 < mean(X1) - stdev(X1) AND X10 < mean(X10) - std(X10)?

```python
def meanMinusStd(x):
    return np.mean(x) - np.std(x)

df2 = df1.where(df1.X1 < meanMinusStd(df1.X1)).where(df1.X10 < meanMinusStd(df1.X10)).dropna()
df2
```

```python
print(f'There are {df2.shape[0]} which companies satisfy the condition.')
```

### What is the ratio of the bankrupted companies among the sub-groups above?

```python
df3 = df2.where(df1.Bankruptcy == 1).dropna()
print(f'The ratio of the bankrupted companies among the sub-groups above is : {df3.shape[0] / df2.shape[0]}')
```
